- 'ECCS-1542152'

We also have the names of known faculty affiliates which can be used as search terms.

If we think of trying to predict relevance to the Stanford Nano shared facilities directly, the problem is quite complex. There simply is not enough data to build a separate model for each researcher and attempting to build one model for the whole facility requires some domain coherence between the publications. In the case of the Critical Zone Observatory data, this domain association existed. In this case, all the researchers use the same facility but their research areas differ. Instead, we can re-frame the problem in terms of identifying publications in our results set from researchers with the same name as one of our approved faculty but who work in different domains. Dimensions has a field called `for_category` that we can use to approximate the research areas covered in Stanford Nano reports. We can find all of the unique `for_category` values in the reported publications then build a training set where the reported publications serve as positive examples and a random selection of publications from categories not covered in the reports can serve as the negative examples. In this way we can simplify the problem. Instead of predicting relevance for each researcher or for the facility as a whole, we only need to predict relevance to these `for_category` values. Of the 2112 publications returned for our name query in 2020, if one of them is associated with a different `for_category`, we assume that it was written by another person with a similar name and discarded. First let's see what categories are covered in the reports.

```python
merged_report = pd.concat([reported_2019, reported_2020])
merged_report.to_csv('../data/merged_report.csv')
```

```python
category_counter = Counter()
for i in merged_report['category_for'].tolist():
    for j in i.strip('][').split(', '):
        if len(j.split(' ')[0]) == 3: # only get the higher level cateogories which are 2 digit numbers
            category_counter.update([j])
```

```python
for i, c in category_counter.items():
    print(f"{i}: {c}")
```

```python
with open('../data/train/train.txt', 'r') as f:
    train, test = train_test_split(f.readlines(), test_size=910)
    # Write training data to a file as required by fasttext
    training_file = open('../data/train/train.train','w')
    training_file.writelines(train)
    training_file.close()
    # Write training data to a file as required by fasttext
    test_file = open('../data/train/train.test','w')
    test_file.writelines(test)
    test_file.close()
```

```python
model = fasttext.train_supervised(input='../data/train/train.train')
```

Now lest test on our validation set. This is just a gut check as we know from other cases that it should work well.

```python
model.test('../data/train/train.valid')
```

Now lets train on the full set and predict relevance of our name search results.

```python
final_model = fasttext.train_supervised(input='../data/train/train.txt')
```

```python
def predict_relevance(features):
    return str(final_model.predict(features)[0]).replace("'__label__", '').replace("',", '').replace('(', '').replace(')', '')
```

```python
name_search_results_2019['concepts'].replace('', np.nan, inplace=True)
name_search_results_2019.dropna(subset=['concepts'], inplace=True)
```

```python
name_search_results_2019['relevance'] = name_search_results_2019.apply(lambda row : predict_relevance(str(row['concepts']).replace("['", "'").replace("']", "'").replace(", ", " ")), axis = 1)
```

```python
# Remove non-relevant publicaitons from the dataframe
name_search_results_2019_filtered = name_search_results_2019[~name_search_results_2019.relevance.str.contains('non-relevant')]
```

```python
print(f"There were {name_search_results_2019.shape[0]} publications before the intervention.")
print(f"There are {name_search_results_2019_filtered.shape[0]} publications after the intervention.")
```

More than half of the publications are predicted to be false positives. Its not clear how to evaluate this fully but we can see how well the name search results cover the reported publications before and after the intervention.

It seems to have worked pretty well. At the cost of removing 12 true positives, we have potentially reduced the false positives by 904. A suggested strategy for this facility is to keep all results returned from querying the names of the institution and the grant numbers. These terms are likely to be highly specific to the facility; it is unlikely that a publication would include any of these terms while not being relevant to the facility. For that reason, all of these publications should be reviewed manually. We can add to these the publications from the name search that our model predicts as relevant. 

```python
search_results_2021 = pd.read_csv('../data/search_results_2021.csv', header=0)
print(f"There were {search_results_2021.shape[0]} publicaitons in 2021 returned from the facility and grant number query.")
```

```python
name_search_results_2021 = pd.read_csv('../data/name_search_results_2021.csv', header=0)
print(f"There were {name_search_results_2021.shape[0]} publicaitons in 2021 returned from the name query.")
name_search_results_2021.drop_duplicates(subset=['doi'], keep='first', inplace=True)
print(f"After duplicates are removed, there are {name_search_results_2021.shape[0]} publicaitons left.")
```

Now lets use our model to predict the relevance of these publications.

```python
name_search_results_2021['relevance'] = name_search_results_2021.apply(lambda row : predict_relevance(str(row['concepts']).replace("['", "'").replace("']", "'").replace(", ", " ")), axis = 1)
```

```python
# Remove non-relevant publicaitons from the dataframe
name_search_results_2021_filtered = name_search_results_2021[~name_search_results_2021.relevance.str.contains('non-relevant')]
```

```python
print(f"After removing the predicted non-relavent publications, there are {name_search_results_2021_filtered.shape[0]} remaining.")
```

Finally, let's merge these name search results with the facility query results, delete any duplicates, and see how many publications are in the final results.

```python
final_2021 = pd.concat([search_results_2021, name_search_results_2021_filtered])
final_2021.drop_duplicates(subset=['doi'], keep='first', inplace=True)
final_2021.to_csv('../data/final_2021.csv')
print(f"There are {final_2021.shape[0]} publications remaining.")
```

This is a fair approximation of how many publications will need to be manually reviewed each year for this facility.
